```python
# Import dependencies
import pandas as pd
```

```python
files2018 = {'Resources/201801.csv','Resources/201802.csv','Resources/201803.csv',
            'Resources/201804.csv','Resources/201805.csv','Resources/201806.csv',
            'Resources/201807.csv','Resources/201808.csv','Resources/201809.csv'}

# Empty lsit
df2018_list =[]

# Loop for each file in the created list
for file in files2018:
    df2018 = pd.read_csv(file)
    df2018_list.append(df2018)

# Concatenate dataframes
df2018 = pd.concat(df2018_list)

df2018
```

```python
# Gender mapping
gendertext = {0:"Unknown", 1:"Male", 2:"Female"}

df2018['gender'] = df2018['gender'].map(gendertext)

df2018
```

```python
df2018.to_csv('Resources/citibike2018.csv', index = False)
```
